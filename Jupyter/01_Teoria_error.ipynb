## 1. Redondeo
Es el proceso de aproximar cifras en la expresión decimal

```
import math as m
from math import pi

x1=(1+m.sqrt(5))/2
x2=m.e
x3=pi

y1=round(x3,3)
y2=m.floor(x3*1000)/1000
y3=m.ceil(x3*1000)/1000

print('Número áureo: ', x1)
print('Número euler: ', x2)
print('Número pi: ', x3)
print('redondear: ', y1)
print('redondear: ', y2)
print('redondear: ', y3)
```

## 2. Truncamiento
Es el proceso de descartar cifras en la expresión decimal

```
# Error de coma flotante
x1=0.1
x2=0.2
X3=x1+x2
x4=pow(m.sqrt(2),2)

y1=x1+x2
y2=m.floor(y1*10)/10
y3=x4
y4=m.floor(y3*10)/10

print('Error coma flotante : ', y1)
print('tratamiento error   : ', y2)
print('Error coma flotante : ', y3)
print('tratamiento error   : ', y4)
```

```
# Truncamiento intencionado
x1=-7.4

y1=int(x1)
y2=m.trunc(x1)
y3=m.floor(x1)

print('Truncar: ', y1)
print('Truncar: ', y2)
print('Truncar: ', y3)
```

## 3. Propagación
Es el proceso de operar cifras aproximadas o truncadas

```
x1=pi
x2=3+4j
x3=m.exp(1)

y1=10*m.sin(2*pi)
y2=abs(x2)
y3=m.log(x3)

print('Aproximación pi: ', x1)
print('Tipo de dato pi: ', type(x1))
print('Propagación Error: ', y1)
print('-------------------')
# complejos
print('Conjugado : ', x2)
print('Tipo de dato conjugado: ', type(x2))
print('Magnitud conjugado: ', y2)
print('Tipo de dato magnitud conjugado: ', type(y2))
print('-------------------')
print('Aproximación e: ', x3)
print('Propagación Error: ', y3)
```
